```python
import os
import json

def write_tmp_json(chord_annotation_dict: dict, out_dir: str, fname: str):
    """
    XXX This function will be modified to output a JAMS-score file.
    """
    #print(os.path.join(out_dir, fname+'.json'))
    with open(os.path.join(out_dir, fname+'.json'), 'w') as fp:
        json.dump(chord_annotation_dict, fp, indent=4)
```
